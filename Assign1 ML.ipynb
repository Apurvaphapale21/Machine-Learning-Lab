```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Create the boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, y='rate')
plt.title('Distribution of Restaurant Ratings')
plt.ylabel('Ratings')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['year'], bins=20, kde=True, color='blue')
plt.title('Year Distribution')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data['location'].value_counts()[:10].plot(kind='bar', color='orange')
plt.title('Top 10 Locations')
plt.xlabel('Location')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data["rate"], bins=10, kde=True, color="blue")
plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data['rate'].astype(str).str.replace('/5', '').astype(float).hist(bins=20, color='purple')
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
cuisine_counts = data["cuisines"].value_counts().nlargest(5)
plt.figure(figsize=(10, 5))
sns.barplot(x=cuisine_counts.index, y=cuisine_counts.values, palette="coolwarm")
plt.title("Top 5 Cuisine Types")
plt.xticks(rotation=45)
plt.xlabel("Cuisine Type")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x="online_order", data=data, palette="pastel")
plt.title("Online Ordering Availability")
plt.xticks(ticks=[0, 1], labels=["No", "Yes"])
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x="book_table", data=data, palette="pastel")
plt.title("Table Booking Availability")
plt.xticks(ticks=[0, 1], labels=["No", "Yes"])
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=data["approx_cost(for two people)"], y=data["rate"], hue=data["location"], palette="viridis", s=100)
plt.title("Cost vs Rating")
plt.xlabel("Cost for Two People")
plt.ylabel("Rating")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=data["votes"], y=data["rate"], hue=data["location"], palette="coolwarm", s=100)
plt.title("Votes vs Rating")
plt.xlabel("Number of Votes")
plt.ylabel("Rating")
plt.show()
```
